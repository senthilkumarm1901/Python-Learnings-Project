###### Basics of Regular Expressions

    What is RegEx? -- Highly useful in Text Data Processing
    An expression containing 'meta' characters and literals to identify and/or replace a pattern matching that expression
        Meta Characters: these characters have a special meaning
        Literals: these characters are the actual characters that are to be matched
    Use Cases
        To Search a string pattern
        To break or split a string based on a pattern
        To replace a part of the string

```python
## Reading Material before you begin executing the below statements
YouTubeVideo('DRR9fOXkfRE')
#Additional YouTube Links:
#https://www.youtube.com/watch?v=FCFdgymqpUI
#https://www.youtube.com/watch?v=lvGtejG69zg&t=3s
```

```python
import re
#the package for Regular Expressions in Python
```

```python
#Most useful functions/methods in re package
#re.match()
#re.search()
#re.findall()
#re.split()
#---Other useful functions you could explore at your convenience---#
#re.compile()
#re.sub()
```

```python
#re.match(pattern, string)
#finds pattern if it occurs at start of the string
re.match(r'Xbox1','Xbox One is better than PS4. But PS4 seems to have better game lineup').group()
#r denotes raw string literal...meaning save the string as it appears... good practice to have it 
```

```python
re.match(r'But','Xbox One is better than PS4. But PS4 seems to have better game lineup').group()
#you need group() method to return the matching string
```

```python
#re.search(pattern, string)
#finds pattern anywhere at the string, but returns only the first instance of the pattern
re.search(r'Xbox','Xbox One is better than PS4. But Xbox1 seems to have better game lineup').group()
```

```python
#re.findall(pattern, string)
#finds pattern anywhere at the string, but returns all the matching patterns
re.findall(r'Xbox','Xbox One is better than PS4. But Xbox1 seems to have better game lineup')
```

```python
#re.split(pattern, string)
re.split(r'\.\s','Xbox One is better than PS4. But PS4 seems to have better game lineup')
```

```python
#Some basic use cases for pattern matching using a few Regex meta or special characters
Input_List=['arun@latentview.co.uk', 'aishwarya@latentview.co.us','bala@gmail.com','banu@yahoo.co.in','yamini@outlook.co.uk']
Ans_List=[]
#to search all email ids begining with a
print "Email Ids beginning with letter a"
for i in range(len(Input_List)):
    if re.search(r'^a',Input_List[i],re.I):
        Ans_List.append(Input_List[i])
print Ans_List
```

```python
Ans_List=[]
#to search all email ids ending with uk
print "Email Ids ending with uk"
for i in range(len(Input_List)):
    if re.search(r'uk$',Input_List[i],re.I):
        Ans_List.append(Input_List[i])
print Ans_List
```

```python
Ans_List=[]
#to search all email ids NOT ending with .com
print "Email Ids ending with uk"
for i in range(len(Input_List)):
    if re.match(r'(?!.*com$)',Input_List[i],re.I):
        Ans_List.append(Input_List[i])
print Ans_List
```

###### Defining Functions

###### A function is a block of organized, reusable code that is used to perform a single, related action. 
#Functions provide better modularity for your application and a high degree of code reusing.
#It is like our own functions similar to built-in functions

```python
#Let us create a function that gets all the domain names from a list of URLs
def domain_name_extraction(url):
    temp=re.sub(r"https://","",url,re.I)
    temp=re.sub(r"www\.","",temp,re.I)
    text=re.search(r'([a-zA-Z\.]+)(\.com|\.org)',temp).group(1)
    return text

List_URLs=["https://www.programiz.com/python-programming/built-in-function","https://worditout.com/word-cloud/2032848","https://docs.python.org/2/howto/unicode.html"]
List_DomainNames=[]
for each in List_URLs:
    List_DomainNames.append(domain_name_extraction(each))
print List_DomainNames

#it uses Regex, lists, and the concept of using reusable modules called functions
```
