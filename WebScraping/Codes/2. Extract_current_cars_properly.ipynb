```python
Owned_vehicle_list=[]
for each in range(len(ip)):
    if ip.loc[each,'current_car']!='EMPTY':
        if re.search(r'(?=.*,)',ip.loc[each,'current_car'],re.I):
            Owned_vehicle_list.extend(ip.loc[each,'current_car'].split(','))
        elif re.search(r'(?=.*;)',ip.loc[each,'current_car'],re.I):
            Owned_vehicle_list.extend(ip.loc[each,'current_car'].split(';'))
        elif re.search(r'(?=.*;)',ip.loc[each,'current_car'],re.I):
            Owned_vehicle_list.extend(ip.loc[each,'current_car'].split(';'))
        elif re.search(r'^(?!.*?(\b\w\/\w\b))(?=.*\/{1,2})',ip.loc[each,'current_car'],re.I):
            #print (ip.loc[each,'current_car'].split(r'/'))
            Owned_vehicle_list.extend(ip.loc[each,'current_car'].split(r'/'))
        elif re.search(r'(?=.*\|)',ip.loc[each,'current_car'],re.I):
            #print (ip.loc[each,'current_car'].split('|'))
            Owned_vehicle_list.extend(ip.loc[each,'current_car'].split('|'))
        elif re.search(r'(?=.*\')',ip.loc[each,'current_car'],re.I):
            if len(re.findall(r"'",ip.loc[each,'current_car'],re.I))>1:
                #print (ip.loc[each,'current_car'].split(r"'"))
                Owned_vehicle_list.extend(ip.loc[each,'current_car'].split(r"'"))
        elif re.search(r'\b\d{2}\b',ip.loc[each,'current_car'],re.I):
            if len(re.findall(r"\b\d{2}\b",ip.loc[each,'current_car'],re.I))>1:
                text=ip.loc[each,'current_car']
                text=re.sub(r'\s(\d{2}\s)',r',\1',text,re.I)
                #print (text.split(','))
                Owned_vehicle_list.extend(text.split(','))
        elif re.search(r'\b\d{4}\b',ip.loc[each,'current_car'],re.I):
            if len(re.findall(r"\b\d{4}\b",ip.loc[each,'current_car'],re.I))>1:
                text=ip.loc[each,'current_car']
                text=re.sub(r'\s(\d{4}\s)',r',\1',text,re.I)
                #print (text.split(','))
                Owned_vehicle_list.extend(text.split(','))
```

```python
Owned_vehicle_list=[each.lower().strip() for each in Owned_vehicle_list if len(each)>1]
Owned_vehicle_list=[re.sub(re.compile(r"'?\b([01]\d)\b",re.I),r'20\1',text) for text in Owned_vehicle_list]
Owned_vehicle_list=[re.sub(re.compile(r"'?\b([56789]\d)\b",re.I),r'19\1',text) for text in Owned_vehicle_list]
```
