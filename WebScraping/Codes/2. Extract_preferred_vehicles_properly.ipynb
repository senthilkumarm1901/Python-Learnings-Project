```python
import pandas as pd
import nltk
import re
```

```python
import pickle
```

```python
ip=pd.read_csv(r'..\Reddit_Comments\Final_reddits_Jul13.csv',index_col=False)
```

```python
ip['Content'].fillna(r'EMPTY',inplace=True)
```

```python
ip['Content']=ip['Content'].str.replace(r"\'",'')
```

```python
considered_vehicle_list=[]
```

```python
considered_vehicle_list=[]
for each in range(len(ip)):
    if each%1==0: print (each)
    text=ip.ix[each,'Content']
    if re.search(r'considered.{0,2}\W+(.*?)(Is.*?vehicle|[.:])',text):
        #print (re.search(r'considered.{0,2}\W+(.*?)(Is.*?vehicle|[.:])',text).group())
        text=re.search(r'considered.{0,2}\W+(.*?)(Is.*?vehicle|[.:])',text).group(1)
        print (text)
        if re.search(r'(?=.*,)(?=.*\()',text): 
            text=re.sub(re.compile(r'(\(.*?\))',re.I),' ',text)
            considered_vehicle_list.extend(text.split(','))
            print ("comma_bracket:",each, text.split(','))
        elif re.search(r'(?=.*,)',text):
            considered_vehicle_list.extend(text.split(','))
            print ("comma_only:",each, text.split(','))
        elif re.search(r'(?=.*;)',text):
            considered_vehicle_list.extend(text.split(';'))
            print ("semi_colon_only:",each, text.split(';'))
        elif re.search(r'(?=.*\')',text):
            if len(re.findall(r"'",text))>1:
                #print (ip.loc[each,'current_car'].split(r"'"))
                considered_vehicle_list.extend(text.split(r"'"))
                print ("apostrophe_only:",each, text.split("'"))
        elif re.search(r'\b\d{2}\b',text):
            if len(re.findall(r"\b\d{2}\b",text))>1:
                text=re.sub(r'\s(\d{2}\s)',r',\1',text,re.I)
                #print (text.split(','))
                considered_vehicle_list.extend(text.split(','))
                print ("YY number_split:",each, text.split(','))
        elif re.search(r'\b\d{4}\b',text):
            if len(re.findall(r"\b\d{4}\b",text))>1:
                text=re.sub(r'\s(\d{4}\W)',r',\1',text,re.I)
                #print (text.split(','))
                considered_vehicle_list.extend(text.split(','))
                print ("YYYY number_split:",each, text.split(','))
        elif re.search(r'([A-Z](\w+)\s){1,}\d?',text):
            text=re.search(r'([A-Z]([A-Za-z0-9\/]+)\s)(([A-Za-z0-9\/]+)\s){1,}\d|([A-Z]([A-Za-z0-9\/]+)\s?){1,}\d?|\b([a-z]+[0-9]+)\b|\b([0-9]+[a-z]+)\b',text).group()
            considered_vehicle_list.append(text)
            print ("plain_vehicle_model:",text)
```
